```python
import random
import matplotlib.pyplot as plt
import numpy as np

# Generate 50 random floating-point numbers between 0 and 1
random_values = [random.random() for _ in range(50)]

# Plot a histogram to visualize the distribution
plt.hist(random_values, bins=10, edgecolor='red')
plt.title("Histogram of 50 Random Floating-Point Numbers")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

# Calculate the mean, median, and standard deviation 
mean_number = np.mean(random_values)
median_number = np.median(random_values)
stdev_number = np.std(random_values)

# Identify the max and min values
max_value = np.max(random_values)
min_value = np.min(random_values)

# Print the calculations
print(f"Mean: {mean_number}")
print(f"Median: {median_number}")
print(f"Standard Deviation: {stdev_number}")
print(f"Maximum Value: {max_value}")
print(f"Minimum Value: {min_value}")
```
